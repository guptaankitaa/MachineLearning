```python
#Seaborn
#Data visuavlization

import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#NOTE = seaborn has some build in dataset
#total bill vs tip dataset

tips = sb.load_dataset('tips')
tips.head()
```

```python
#visuvalize the data
sb.set_theme()
sb.relplot(data=tips, x= 'total_bill', y='tip', col='time',hue='smoker',style='smoker',size='size')
```

```python
# iris dataset

iris = sb.load_dataset('iris')
iris.head()
```

```python
#SCATTER PLOT

sb.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sb.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
#loading the titanic dataset

titanic = sb.load_dataset('titanic')
titanic.head()
titanic.shape
```

```python
#count plot

sb.countplot(x='class',data=titanic)
```

```python
sb.countplot(x='survived',data=titanic)
```

```python
#BAR CHART

sb.barplot(x='sex', y='survived',hue='class',data=titanic)
```
