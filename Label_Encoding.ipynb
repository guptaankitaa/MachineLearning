convet the labels into numeri form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
cancer_data = pd.read_csv('/content/breast_cancer_data.csv')
```

```python
cancer_data.head()
```

```python
#finding the count of diff labels
cancer_data['diagnosis'].value_counts()
```

```python
# load the label encoder function
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_data.diagnosis)
```

```python
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

```python
cancer_data['target'].value_counts()
```
